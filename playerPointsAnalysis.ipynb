```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/box_scores10242023')
```

```python
df.head()
```

```python
df['bins'] = pd.cut(df['PTS'],
					bins = [0,10,20,30,40,50,100], 
					labels = ["0-10", "11-20", "21-30", "31-40", "41-50", "50+"])

a = df.groupby('bins', observed = True).size()
a.plot.pie(figsize=(4,4), autopct='%1.0f%%')
plt.title("Player Points")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df['MP'], df['PTS'])
plt.title("Minutes Played vs Field Goal Percentage")
plt.xlabel("Minutes Played")
plt.ylabel("Number of Points")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df['FG%'], df['PTS'])
plt.title("Player Points vs Field Goal Percentage")
plt.xlabel("Field Goal %")
plt.ylabel("Number of Points")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df['FGA'], df['PTS'])
plt.title("Player Points vs Field Goal Attempts")
plt.xlabel("Number of Field Goal Attempts")
plt.ylabel("Number of Points")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df['ORtg'], df['PTS'])
plt.title("Player Points vs Offensive Rating")
plt.xlabel("Offensive Rating")
plt.ylabel("Number of Points")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df['ORtg'], df['FG%'])
plt.title("Offensive Rating vs Field Goal Percentage")
plt.xlabel("Offensive Rating")
plt.ylabel("Field Goal %")
plt.show()
```
